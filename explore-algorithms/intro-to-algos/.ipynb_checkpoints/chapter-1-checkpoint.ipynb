1.2-2 Behaviour of merge (64n lg(n)) vs insertion sort $8n^{2}$

```python
import numpy as np
from matplotlib import pyplot as plt

# Fake Data 
x = range(1,100)
insert_sort = 8 * np.power(x, 2)
merge_sort = 64 * np.multiply(x, np.log(x))

plt.plot(x, insert_sort, "-", label="$8n^{2}$")
plt.plot(x, merge_sort, "-", label="x ln(x)")

plt.legend()
plt.show()
```

Until x = 25 insertion sort performs better, beyond that merge is better.

Insertion sort Sample - non decreasing

```python
input = [5, 2, 4, 6, 1, 3]
for j in range(1, len(input)):
    key = input[j]
    i = j - 1
    while (i > -1 and input[i] > key):
        input[i+1] = input[i]
        i = i - 1
    input[i + 1] = key
print(input)
```

__2.1-2__ Rewrite Insertion-Sort into non-increasing from non-decreasing?

```python
input = [5, 2, 4, 6, 1, 3,1]
cost = 0
for j in range(1, len(input)):
    cost = cost + 1
    key = input[j]
    i = j - 1
    while (i > -1) and (input[i] < key):
        cost = cost + 1
        input[i+1] = input[i]
        i = i -1
    input[i+1] = key
print(input)
print(cost)
```
